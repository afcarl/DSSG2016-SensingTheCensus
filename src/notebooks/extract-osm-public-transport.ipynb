## Import packages and read file

```python
import pandas as pd
from pyquery import PyQuery as pq
from lxml import etree
```

```python
osm = pq(filename="../../data/OSM/planet_8.623,45.245_9.656,45.712.osm")
```

## Get all nodes corresponding to amenities 

```python
pt_nodes = osm("tag[k='public_transport']").parents("node")
```

```python
pt_dict=[]

for node in pt_nodes:
    pt_dict.append({'id' : node.attrib["id"],
                       'lon': node.attrib["lon"],
                       'lat': node.attrib["lat"],
                        node.find(".//tag[@k='public_transport']").attrib['k'] : node.find(".//tag[@k='public_transport']").attrib['v'],
                        })
```

```python
pt_df = pd.DataFrame(pt_dict)
```
