### This is a script that reads Region-to-Region network data and aggregate them as a single file.

```python
import pandas as pd
import json
import numpy as np
from pandas.io.json import json_normalize
from scipy.spatial.distance import cosine
import csv
import ast
import os
```

```python
alldf = pd.DataFrame(columns=['month','day','hour','source','dest','call'])

for month in {'11','12'}:
    for day in range(1,32):        
        to_read = '../../data/CDR/region-network/ACEtoACE-2013-' + month + '-' + \
                    str(day).zfill(2) + '.csv'
        
        if os.path.isfile(to_read):
            # loading the raw data
            df = pd.read_csv(to_read) 
            df.columns = ['source','dest','hour','call']
            df['month'] = int(month)
            df['day'] = day
            
            alldf = alldf.append(df)

```

```python
alldf = alldf[['month','day','hour','source','dest','call']]
alldf.head()
```

```python
output_filename = '../../data/CDR/generated/region_network_final.csv'
alldf.to_csv(output_filename, encoding='utf-8', index=False)
```
