```python
df_list_income = []
df_list_census = []
df_list_income_detailed = []
for procom, gb_df in pd.groupby(df_census, "PROCOM"):
    if not np.isnan(procom):
        df_income_procom = df_income[df_income['Codice Istat Comune'] == procom]
        income_histogram = df_income_procom[['Reddito complessivo minore di zero euro - Frequenza',
                                             'Reddito complessivo minore di zero euro - Ammontare',
                                             'Reddito complessivo da 0 a 10000 euro - Frequenza',
                                             'Reddito complessivo da 0 a 10000 euro - Ammontare',
                                             'Reddito complessivo da 10000 a 15000 euro - Frequenza',
                                             'Reddito complessivo da 10000 a 15000 euro - Ammontare',
                                             'Reddito complessivo da 15000 a 26000 euro - Frequenza',
                                             'Reddito complessivo da 15000 a 26000 euro - Ammontare',
                                             'Reddito complessivo da 26000 a 55000 euro - Frequenza',
                                             'Reddito complessivo da 26000 a 55000 euro - Ammontare',
                                             'Reddito complessivo da 55000 a 75000 euro - Frequenza',
                                             'Reddito complessivo da 55000 a 75000 euro - Ammontare',
                                             'Reddito complessivo da 75000 a 120000 euro - Frequenza',
                                             'Reddito complessivo da 75000 a 120000 euro - Ammontare',
                                             'Reddito complessivo oltre 120000 euro - Frequenza',
                                             'Reddito complessivo oltre 120000 euro - Ammontare']]

        ih_freq = income_histogram[['Reddito complessivo minore di zero euro - Frequenza',
                       'Reddito complessivo da 0 a 10000 euro - Frequenza',
                       'Reddito complessivo da 10000 a 15000 euro - Frequenza',
                       'Reddito complessivo da 15000 a 26000 euro - Frequenza',
                       'Reddito complessivo da 26000 a 55000 euro - Frequenza',
                       'Reddito complessivo da 55000 a 75000 euro - Frequenza',
                       'Reddito complessivo da 75000 a 120000 euro - Frequenza',
                       'Reddito complessivo oltre 120000 euro - Frequenza']].values.squeeze()

        ih_sum = income_histogram[['Reddito complessivo minore di zero euro - Ammontare',
                       'Reddito complessivo da 0 a 10000 euro - Ammontare',
                       'Reddito complessivo da 10000 a 15000 euro - Ammontare',
                       'Reddito complessivo da 15000 a 26000 euro - Ammontare',
                       'Reddito complessivo da 26000 a 55000 euro - Ammontare',
                       'Reddito complessivo da 55000 a 75000 euro - Ammontare',
                       'Reddito complessivo da 75000 a 120000 euro - Ammontare',
                       'Reddito complessivo oltre 120000 euro - Ammontare']].values.squeeze()

        cs = np.cumsum(ih_freq)
        idx = np.argmin(np.abs(cs - cs[-1]/2)) + 1
        mean_med_bin = ih_sum[idx] / ih_freq[idx]
        df_list_income.append(mean_med_bin)

        this_df_income = pd.DataFrame({'Reddito da fabbricati': df_income_procom['Reddito da fabbricati - Ammontare'] / df_income_procom['Reddito da fabbricati - Frequenza'],
                   'Reddito da lavoro dipendente e assimilati':df_income_procom['Reddito da lavoro dipendente e assimilati - Ammontare']/df_income_procom['Reddito da lavoro dipendente e assimilati - Frequenza'],
                   'Reddito da pensione':df_income_procom['Reddito da pensione - Frequenza']/df_income_procom['Reddito da pensione - Ammontare'],
                   'Reddito da lavoro autonomo (comprensivo dei valori nulli)': (df_income_procom['Reddito da lavoro autonomo (comprensivo dei valori nulli) - Ammontare'] / 
                                                                                df_income_procom['Reddito da lavoro autonomo (comprensivo dei valori nulli) - Frequenza']),
                   "Reddito di spettanza dell'imprenditore in contabilita' ordinaria  (comprensivo dei valori nulli)": (df_income_procom["Reddito di spettanza dell'imprenditore in contabilita' ordinaria  (comprensivo dei valori nulli) - Ammontare"] / 
                                                                                   df_income_procom["Reddito di spettanza dell'imprenditore in contabilita' ordinaria  (comprensivo dei valori nulli) - Frequenza"]),
                   "Reddito di spettanza dell'imprenditore in contabilita' semplificata  (comprensivo dei valori nulli)": (df_income_procom["Reddito di spettanza dell'imprenditore in contabilita' semplificata (comprensivo dei valori nulli) - Ammontare"] / 
                                                                                   df_income_procom["Reddito di spettanza dell'imprenditore in contabilita' semplificata (comprensivo dei valori nulli) - Frequenza"])
                                      })

        
        df_list_income_detailed.append(this_df_income)
        
        sgb = gb_df.sum()
        population = sgb["P1"]
        high_school = sgb["P48"]/sgb["P1"]
        illiteracy = sgb["P52"]/sgb["P1"]
        sixtyfive_plus = (sgb["P27"]+sgb["P28"]+sgb["P29"])/sgb["P1"]
        foreigners = sgb["ST15"]/sgb["P1"]
        rented_dwelling = sgb["A46"]/(sgb["A46"]+ sgb["A47"]+ sgb["A48"])
        unemployment = sgb["P61"]/sgb["P60"]
        df_list_census.append(pd.DataFrame({
                        "procom":procom,
                        "population":population,
                        "high_school":high_school,
                        "illiteracy":illiteracy,
                        "sixtyfive_plus":sixtyfive_plus,
                        "foreigners":foreigners,
                        "rented_dwelling":rented_dwelling, 
                        "unemployment":unemployment}, index=[0]))
        
```

```python
df_census_sub = pd.concat(df_list_census)
df_income_sub = pd.concat(df_list_income_detailed)
```

```python
import sklearn.linear_model as lm
from sklearn.kernel_ridge import KernelRidge
from sklearn.cross_validation import KFold
from sklearn.metrics import r2_score
```

```python
def run_cross_validation(X, y, reg_class, **kwargs):
    # Construct a kfolds object
    kf = KFold(len(y), n_folds=5, shuffle=True)
    y_pred = y.copy()
    
    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        # Initialize a classifier with key word arguments
        clf = reg_class(**kwargs)
        clf.fit(X_train, y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred

#def accuracy(y_true, y_pred):
#    # NumPy inteprets True and False as 1. and 0.
#    return np.mean(y_true == y_pred)lm.ElasticNet()
```

```python
pred_y = run_cross_validation(X, y, lm.LinearRegression)
```
