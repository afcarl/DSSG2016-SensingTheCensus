## Import packages and read file

```python
import pandas as pd
from pyquery import PyQuery as pq
from lxml import etree
```

```python
osm = pq(filename="../../data/OSM/planet_8.623,45.245_9.656,45.712.osm")
```

## Get all nodes corresponding to amenities 

```python
amenity_nodes = osm("tag[k='amenity']").parents("node")
```

```python
amenity_dict=[]

for node in amenity_nodes:
    amenity_dict.append({'id' : node.attrib["id"],
                       'lon': node.attrib["lon"],
                       'lat': node.attrib["lat"],
                        node.find(".//tag[@k='amenity']").attrib['k'] : node.find(".//tag[@k='amenity']").attrib['v'],
                        })
```

```python
amenity_df = pd.DataFrame(amenity_dict)
```
