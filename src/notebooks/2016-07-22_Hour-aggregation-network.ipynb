```python
import pandas as pd
import json
import numpy as np
from pandas.io.json import json_normalize
from scipy.spatial.distance import cosine
import csv
import ast
import os
```

```python
tp = pd.read_csv('../../data/CDR/generated/region_network_MItoMI-2013-12-26.txt') 

```

```python
min(tp.time)
```

```python
tp.index = pd.to_datetime(tp['time'],unit='ms',utc=True)
tp.index = tp.index.tz_localize('UTC').tz_convert('Europe/Rome')
tp['date'] = tp.index
tp['time_hour'] = tp.index.hour
tp['month'] = tp.index.month
tp['day'] = tp.index.day
```

```python
tp = tp.sort_values(['time'])
```

```python
tp.head()
```

```python
final = tp.groupby(['source_region','dest_region','month','day','time_hour'], as_index=False).agg({     
            'time': 'min',
            'date': 'min',
            'adjusted_call': 'sum'            
        })
```

```python
final = final.sort_values(['source_region','dest_region','time'])
```
